To check if the stripplot and regplot would have given the same results, so that we can use stripplot for everything.

```python
sns.regplot(x = x_col[2], y = 'FamilyIncome', data = data)
```

Lmplot also gives us the same results

```python
sns.lmplot(x = x_col[2], y = 'FamilyIncome', data = data)
```
