```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('new_data.csv')
```

```python
data.head()
```

```python
data = data.drop(['Unnamed: 0'], axis = 1)
```

From Model 4, we found some variables we had to drop because of multicollinearity and not enough correlation between independent and dependent variables.

Acres, YearBuilt, Language, HeatingFuel, NumChildren, NumBedrooms are the variables we will drop here as well.

```python
data = data.drop(['Acres', 'YearBuilt', 'Language', 'HeatingFuel', 'NumChildren', 'NumBedrooms'], axis = 1)
```

```python
data.head()
```

```python
y = data.iloc[:, 0:1]
x = data.iloc[:, 1:]
```

Getting dummy variables after removing dummy variable trap

```python
encoded_x = pd.get_dummies(x, drop_first = True)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(encoded_x, y, test_size = 0.2, random_state = 0)
```

```python
import statsmodels.api as sm
```

Simple multiple regression

```python
X = X_train
Y = Y_train
## fit a OLS model with intercept on TV and Radio
X = sm.add_constant(X)
est = sm.OLS(Y, X).fit()

est.summary()
```

We will try to see if using PCA improves our score.

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = None)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)
explained_variance = pca.explained_variance_ratio_
```

```python
explained_variance
```

```python
pca = PCA(n_components = 2)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)
explained_variance = pca.explained_variance_ratio_
```

```python
explained_variance.sum()
```

```python
import statsmodels.api as sm
```

```python
X = X_train_pca
Y = Y_train
## fit a OLS model with intercept on TV and Radio
X = sm.add_constant(X)
est = sm.OLS(Y, X).fit()

est.summary()
```

PCA is known to work better if values are standardized.

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train_std = sc.fit_transform(X_train)
X_test_std = sc.transform(X_test)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = None)
X_train_pca = pca.fit_transform(X_train_std)
X_test_pca = pca.transform(X_test_std)
explained_variance = pca.explained_variance_ratio_
```

```python
explained_variance
```

```python
pca = PCA(n_components = 8)
X_train_pca = pca.fit_transform(X_train_std)
X_test_pca = pca.transform(X_test_std)
explained_variance = pca.explained_variance_ratio_
```

```python
explained_variance.sum()
```

```python
import statsmodels.api as sm
```

```python
X = X_train_pca
Y = Y_train
## fit a OLS model with intercept on TV and Radio
X = sm.add_constant(X)
est = sm.OLS(Y, X).fit()

est.summary()
```

FactorAnalysis is used if there are mixed variables as in our dataset.

```python
from sklearn.decomposition import FactorAnalysis
```

```python
fa = FactorAnalysis()
X_train_fa = fa.fit_transform(X_train)
X_test_fa = fa.transform(X_test)
```

```python
pd.DataFrame(fa.components_)
```

```python
X = X_train_fa
Y = Y_train
## fit a OLS model with intercept on TV and Radio
X = sm.add_constant(X)
est = sm.OLS(Y, X).fit()

est.summary()
```

FactorAnalysis works better with Standardized variables

```python
fa = FactorAnalysis()
X_train_fa = fa.fit_transform(X_train_std)
X_test_fa = fa.transform(X_test_std)
```

```python
X = X_train_fa
Y = Y_train
## fit a OLS model with intercept on TV and Radio
X = sm.add_constant(X)
est = sm.OLS(Y, X).fit()

est.summary()
```
